# Polynomial Regression
first order : y = mx + b

second order : y = θx² + θx + θ

third order : y = θx³ + θx² + θx + θ

and so on.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)

x = np.random.normal(5.0,1.0,400)
y = np.random.normal(50.0,2.0,400) / x


plt.scatter(x,y, alpha=0.2)
plt.show()
```

## np.polyfit( )

```python
first_order = np.poly1d(np.polyfit(x, y, 1))
second_order = np.poly1d(np.polyfit(x, y, 2))
third_order = np.poly1d(np.polyfit(x, y, 3))
fourth_order = np.poly1d(np.polyfit(x, y, 4))
```

```python
poly_line = np.linspace(2, 9, 100)
plt.scatter(x, y, alpha=0.2)

plt.plot(poly_line, first_order(poly_line), c='y')
plt.plot(poly_line, second_order(poly_line), c='r')
plt.plot(poly_line, third_order(poly_line), c='g')
plt.plot(poly_line, fourth_order(poly_line), c='b')

plt.show()
```

## r-squared :

```python
from sklearn.metrics import r2_score

r2_score(y, fourth_order(x))
```

```python
plt.scatter(x, y, alpha=0.2)
plt.plot(poly_line, fourth_order(poly_line), c='b')
plt.show()
```

```python
def predict_y_for(x, order):
    theta = fourth_order 
    b = theta[0]
    y_pred = 0
    for n in range(order, 0, -1):
        y_pred += theta[n] * (x**n)
    y_pred = y_pred + b
    return y_pred
```

```python
plt.scatter(x, predict_y_for(x, 4), c='r')
plt.plot(poly_line, fourth_order(poly_line), c='b')
plt.show()
```
