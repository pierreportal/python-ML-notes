# Linear Regression & Gradient Descent with Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

### We will train our model on this set of data :

```python
x = np.random.normal(5.0, 1.0, 200)
y = x + np.random.normal(5.0, 2.0, 200) / 5
plt.scatter(x,y,c='g', marker='^')
plt.show()
```

### Create the model :

```python
class LinearRegression(object):
    def __init__(self,x,y):
        self.x = x
        self.y = y
        self.m = len(x)
        self.X = np.ones([self.m,2])
        for i in range(self.m):
            self.X[i][1] = self.x[i]
        self.theta = np.random.randn(2)
     
    def J(self):
        self.h = np.dot(self.X,self.theta)
        return 1/(2*self.m) * sum((self.h - self.y)**2)
     
    def gradient_descent(self,alpha,iterations):
        temp = np.zeros(len(self.theta))
        for i in range(iterations):
            for j in range(len(self.theta)):
                temp[j] = self.theta[j] - alpha * 1/self.m * sum((self.h - self.y)*self.X[:,j])    
            self.theta = temp
            result = self.J()
        print('After %s iterations, J = %s, Theta = %s'%(iterations, result, self.theta))
        plt.scatter(self.X[:,1],self.y,c='g', marker='^')
        plt.plot(self.X[:,1],self.h,c='coral')
        plt.show()
        return self.theta
     
    def predict(self,new_x):
        new_h = self.theta[0] + self.theta[1] * new_x
        plt.scatter(self.X[:,1],self.y, c='g', marker='^',alpha = 0.2)
        plt.scatter(new_x,new_h,c='r')
        plt.plot(self.X[:,1],self.h,c='coral')
        plt.show()
        return new_h
```

### Fit :

```python
model = LinearRegression(x,y)
```

### Gradient Descent :
Let's try 1000 iterations with a learning rate (alpha) of 0.01.

```python
model.gradient_descent(0.01,1000)
```
